### 1. What is the difference between interpreted and compiled languages?

**Compiled languages** are converted into machine code before execution (e.g., C, C++).
**Interpreted languages** are executed line by line by an interpreter (e.g., Python).

### 2. What is exception handling in Python?

Exception handling is a method used to handle runtime errors using `try`, `except`, `else`, and `finally` blocks so the program does not crash.

---

### 3. What is the purpose of the `finally` block in exception handling?

The `finally` block always executes whether an exception occurs or not. It is commonly used to close files or release resources.

---

### 4. What is logging in Python?

Logging is a way to record events, errors, or messages that occur during program execution using the `logging` module.

---

### 5. What is the significance of the `__del__` method in Python?

The `__del__` method is a destructor that is automatically called when an object is destroyed. It is used to release resources.

---

### 6. What is the difference between `import` and `from ... import` in Python?

`import module` imports the whole module.
`from module import name` imports only a specific function, class, or variable from the module.

````python
18) How would you implement a basic logging setup that logs to a file with rotation after 1MB?

import logging
from logging.handlers import RotatingFileHandler

# Create logger
logger = logging.getLogger("MyLogger")
logger.setLevel(logging.INFO)

# Create rotating file handler (1MB limit)
handler = RotatingFileHandler("app.log", maxBytes=1024*1024, backupCount=3)

# Create log format
formatter = logging.Formatter("%(asctime)s - %(levelname)s - %(message)s")
handler.setFormatter(formatter)

# Add handler to logger
logger.addHandler(handler)

# Example log messages
logger.info("Program started")
logger.warning("This is a warning message")
logger.error("This is an error message")

#Explanation:
This program uses the RotatingFileHandler from the logging module to write logs to app.log. When the file size reaches 1MB, it automatically creates a new log file and keeps up to 3 backup files.

````

````python
22)How can you check if a file is empty before attempting to read its contents.

import os

filename = "sample.txt"

if os.path.getsize(filename) == 0:
    print("The file is empty.")
else:
    with open(filename, "r") as file:
        content = file.read()
        print(content)

#Explanation:
The os.path.getsize() function checks the size of the file. If the size is 0 bytes, the file is empty. Otherwise, the program opens the file and reads its contents.

````
